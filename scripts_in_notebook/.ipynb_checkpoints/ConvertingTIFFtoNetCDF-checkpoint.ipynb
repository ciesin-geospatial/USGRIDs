```python
import os
import xarray as xr
import rasterio
import glob

def tif_to_dataset(tif_file):
    # Read TIFF file using rasterio
    with rasterio.open(tif_file) as src:
        # Read raster data
        data = src.read(1)  # Assuming single band TIFF
        # Extract metadata
        metadata = src.meta
    
    # Create xarray DataArray
    da = xr.DataArray(data, dims=('y', 'x'), name=os.path.basename(tif_file).split('.')[0], attrs=metadata)
    
    return da

def create_nc_file(tif_files, output_nc_file):
    # Create an empty xarray Dataset
    ds = xr.Dataset()

    # Iterate over TIFF files and add each variable to the Dataset
    for tif_file in tif_files:
        variable_name = os.path.basename(tif_file).split('.')[0]
        ds[variable_name] = tif_to_dataset(tif_file)

    # Write the Dataset to a NetCDF file
    ds.to_netcdf(output_nc_file)

# Example usage:
tif_path = "G:/usgrids2020/raster/DE/"
tif_list = os.listdir(tif_path)
filenames[1][25:-4]
tif_files = [field[25:-4] for field in filenames]
output_nc_file = 'G:/usgrids2020/output.nc'
create_nc_file(tif_files, output_nc_file)
```
