```python
import arcpy
import os 

arcpy.env.workspace = "G:/usgrids2020/test/"
gdbs = arcpy.ListWorkspaces("*", "FILEGDB")
procList = [os.path.join(gdb,f"{os.path.basename(gdb)[:-4]}_joined") for gdb in gdbs]
```

```python

#Calculate the area(land+water) by GEODESIC method => Kytt recommended
arcpy.management.AddField(procList[0], "GEODESIC_AREA", "FLOAT")
arcpy.management.CalculateGeometryAttributes(procList[0], "GEODESIC_AREA AREA_GEODESIC", "", "SQUARE_METERS")
```

```python
with arcpy.da.SearchCursor(procList[0], ["GEODESIC_AREA"]) as cursor:
    for i in range(10):
        for row in cursor:
            print(row)
```

```python
for field in arcpy.ListFields(procList[0]):
    print(field.name)
```
